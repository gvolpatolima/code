```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,20)
y = np.linspace(-5,5,20)
(x,y)=np.meshgrid(x,y)
u = x
v = -y
plt.xlabel('x')
plt.ylabel('y')
plt.quiver(x,y,u,v)
plt.show
```

```python
plt.streamplot(x,y,u,v,density=7) 
plt.show()
```

```python
x_range = y_range = np.linspace(0,2*np.pi, 200)
x,y = np.meshgrid(x_range, y_range)
fx, fy = np.sin(x), np.cos(y)
f_mag = np.sqrt(fx**2 + fy**2)

plt.figure(figsize=(8,8))
plt.imshow(f_mag, origin='lower', extent=(0,2*np.pi,0,2*np.pi), cmap="gray")
plt.quiver(x[5::10,::10],y[5::10,::10],fx[5::10,::10],fy[5::10,::10],pivot="middle",color="red")
plt.show

```
